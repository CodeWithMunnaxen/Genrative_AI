# Bag of Words (BoW)

- Bag of Words (BoW) is a technique used to convert text into numerical form so that machine learning models can understand it.

- The word "Bag" means:

- Word order is ignored.
- Only the presence or frequency of words matters.

| Property               | Meaning                     |
| ---------------------- | --------------------------- |
| Numeric Representation | Text → Numbers              |
| Frequency Based        | Counts words                |
| Order Ignored          | Position doesn't matter     |
| Fixed Length           | Same vector size            |
| Sparse Matrix          | Many zeros                  |
| Simple                 | Easy implementation         |
| Vocabulary Based       | Depends on unique words     |
| No Semantics           | Doesn't understand meaning  |
| Fast                   | Quick computation           |
| Traditional NLP        | Used before Embeddings/LLMs |


| Advantage                | Explanation                                           | Example                        |
| ------------------------ | ----------------------------------------------------- | ------------------------------ |
| Simple                   | Easy to understand and implement                      | Count words and create vectors |
| Fast                     | Computation is very quick                             | Suitable for small datasets    |
| Easy Conversion          | Converts text into numbers                            | ML models need numeric data    |
| Good Baseline            | Works well as a starting model                        | Text Classification            |
| No Training Required     | Just count words                                      | Unlike embeddings              |
| Works with ML Algorithms | Compatible with Naive Bayes, Logistic Regression, SVM | Spam Detection                 |
| Easy Feature Extraction  | Words become features automatically                   | Vocabulary columns             |


| Disadvantage                      | Explanation                        | Example                           |
| --------------------------------- | ---------------------------------- | --------------------------------- |
| Ignores Word Order                | Position of words is lost          | "Dog bites man" = "Man bites dog" |
| No Meaning Understanding          | Cannot understand semantics        | good ≠ excellent                  |
| Sparse Matrix                     | Lots of zeros                      | Large vocabulary                  |
| High Dimensionality               | More unique words → more columns   | 10,000 words = 10,000 features    |
| Vocabulary Dependency             | New words are ignored              | hockey not in training data       |
| Context Ignored                   | Cannot understand sentence context | "bank" has multiple meanings      |
| Storage Cost                      | Large matrices consume memory      | Huge datasets                     |
| Similar Words Treated Differently | Synonyms not recognized            | car ≠ automobile                  |


```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# Documents
docs = [
    "Munna likes cricket",
    "Raunak likes football",
    "Kuldeep likes cricket",
    "Rohan plays football",
    "Sohan likes cricket"
]

# Create BoW
cv = CountVectorizer()

bow = cv.fit_transform(docs)

# Sparse Matrix -> Dense Matrix
dense = bow.todense()

# Dense Matrix -> List
dense_list = dense.tolist()

# DataFrame
df = pd.DataFrame(
    dense_list,
    columns=cv.get_feature_names_out(),
    index=[f"Doc{i}" for i in range(1, len(docs)+1)]
)

print(df)
```
