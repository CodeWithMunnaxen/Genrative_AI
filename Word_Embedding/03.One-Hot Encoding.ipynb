# One-Hot Encoding
- One-Hot Encoding is a technique used to convert categorical (text) data into numerical format so Machine Learning models can understand it.

# Advantages of One-Hot Encoding

| Advantage                      | Explanation                          |
| ------------------------------ | ------------------------------------ |
| No Order Assumption            | Doesn't create fake ranking          |
| Easy to Understand             | Simple binary representation         |
| Works Well with ML Models      | Logistic Regression, SVM, etc.       |
| Preserves Category Information | Each category gets its own column    |
| Improves Accuracy              | Better than arbitrary numeric labels |


# Disadvantages of One-Hot Encoding

| Disadvantage        | Explanation                                   |
| ------------------- | --------------------------------------------- |
| High Dimensionality | Too many columns if categories are many       |
| Memory Usage        | More RAM required                             |
| Sparse Matrix       | Mostly 0 values                               |
| Slower Training     | Large datasets increase computation           |
| Not Good for NLP    | Thousands of unique words create huge vectors |


# 

| Feature              | One-Hot Encoding | Word Embedding |
| -------------------- | ---------------- | -------------- |

# Using Scikit-Learn

```python
from sklearn.preprocessing import OneHotEncoder

data = [["Red"], ["Blue"], ["Green"]]

encoder = OneHotEncoder()

result = encoder.fit_transform(data)
```

```python
import pandas as pd

# Sample Data
df = pd.DataFrame({
    "Fruit": ["Apple", "Banana", "Mango", "Apple"]
})

print("Original Data:")
print(df)

# One-Hot Encoding
encoded_df = pd.get_dummies(df, columns=["Fruit"])
```

```python
encoded_df
```

```python
from sklearn.preprocessing import OneHotEncoder

# Data
data = [
    ["Apple"],
    ["Banana"],
    ["Mango"],
    ["Apple"]
]

# Create Encoder
encoder = OneHotEncoder(sparse_output=False)

# Transform Data
encoded = encoder.fit_transform(data)
```

```python
encoded
```

```python
from sklearn.preprocessing import OneHotEncoder

data = [
    ["Apple"],
    ["Banana"],
    ["Mango"]
]

encoder = OneHotEncoder(sparse_output=False)

encoded = encoder.fit_transform(data)

print("Categories:")
print(encoder.categories_)

print("\nEncoded Data:")
```

```python
encoded
```

```python
import pandas as pd

# Student Branch Data
df = pd.DataFrame({
    "Student": ["Rahul", "Priya", "Amit"],
    "Branch": ["CSE", "ECE", "MECH"]
})

print("Original Data")
print(df)

# One-Hot Encoding
encoded = pd.get_dummies(df, columns=["Branch"])

print("\nAfter One-Hot Encoding")
print(encoded)
```

| Advantages                                                    | Disadvantages                                                 |
| ------------------------------------------------------------- | ------------------------------------------------------------- |
| Easy to understand and implement                              | Creates many new columns when categories are large            |
| Prevents false ordering between categories                    | Increases memory usage                                        |
| Works well with most Machine Learning algorithms              | Produces sparse matrices (mostly 0 values)                    |
| Preserves all category information                            | Training becomes slower with high-dimensional data            |
| Suitable for nominal data (e.g., Color, City, Gender)         | Not suitable for datasets with thousands of unique categories |
| No category is considered greater or smaller than another     | Cannot capture relationships between categories               |
| Improves model performance compared to simple label encoding  | Feature space grows rapidly as categories increase            |
| Widely supported in ML libraries like Pandas and Scikit-Learn | Not efficient for NLP vocabulary with thousands of words      |
