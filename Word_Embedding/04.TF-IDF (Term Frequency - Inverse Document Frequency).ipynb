```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

docs = [
    "Munna likes cricket cricket",
    "Raunak likes football",
    "Kuldeep likes cricket",
    "Rohan plays football",
    "Sohan likes cricket"
]

tfidf = TfidfVectorizer()

x = tfidf.fit_transform(docs)

print("Vocabulary")
print(tfidf.get_feature_names_out())

print("\nShape")
print(x.shape)

print("\nSparse Matrix")
print(x)

dense = x.todense()

dense_list = dense.tolist()

df = pd.DataFrame(
    dense_list,
    columns=tfidf.get_feature_names_out(),
    index=["Doc1","Doc2","Doc3","Doc4","Doc5"]
)

print("\nTF-IDF DataFrame")
print(df)
```

| Method                    | Meaning                        |
| ------------------------- | ------------------------------ |
| `TfidfVectorizer()`       | TF-IDF object create karta hai |
| `fit()`                   | Vocabulary learn karta hai     |
| `transform()`             | TF-IDF calculate karta hai     |
| `fit_transform()`         | Dono ek sath                   |
| `get_feature_names_out()` | Vocabulary deta hai            |
| `shape`                   | Documents × Features           |
| `todense()`               | Sparse → Dense                 |
| `toarray()`               | Sparse → Numpy Array           |
| `DataFrame()`             | Table format me dikhata hai    |


```python
# fit()           -> Learn Words
# transform()     -> Convert Words to Numbers
# TF              -> Word Frequency in Document
# IDF             -> Word Rarity Across Documents
# TF-IDF          -> Important Word Score
```
