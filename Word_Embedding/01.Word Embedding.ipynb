# Word Embedding
- Word Embedding is a technique that converts words into numerical vectors (numbers) so that machines can understand their meaning and relationships.

# 

# 

# 

| Word     | Vector          |
| -------- | --------------- |
| King     | [0.2, 0.8, 0.4] |
| Queen    | [0.3, 0.7, 0.5] |
| Prince   | [0.1, 0.9, 0.3] |
| Princess | [0.2, 0.8, 0.4] |


# 

# 

# Word Embeddings are needed to convert words into meaningful numerical vectors so that AI models can understand semantic relationships between words and perform tasks like search, translation, classification, RAG, and text generation
