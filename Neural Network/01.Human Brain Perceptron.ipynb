# Human Brain is a biological intelligence system, while a Perceptron is its simplified mathematical model that acts as the basic building block of Artificial Intelligence and Neural Networks.

# What is a Perceptron?

A Perceptron is the simplest Artificial Neural Network (ANN) and acts like a single artificial neuron.

It takes inputs, performs calculations, and produces an output.

# Components of a Perceptron

1. Input Layer (x)

- These are the features or values given to the perceptron.

2. Weights (w)

- Weights represent the importance of each input.

3. Bias (b)

- Bias shifts the decision boundary.

4. Summation Unit (Σ)

- It calculates the weighted sum of all inputs.

5. Activation Function

- The activation function decides whether the neuron should fire or not.

6. Output (y)

- Final prediction.

```python
x1 = 80
x2 = 140
x3 = 4

w1 = 0.5
w2 = 0.3
w3 = 0.2

b = -50

z = (x1*w1) + (x2*w2) + (x3*w3) + b

if z >= 0:
    y = 1
else:
    y = 0

print("Weighted Sum:", z)
print("Output:", y)
```

```python
# Input Data
#     ↓
# Multiply by Weights
#     ↓
# Add Bias
#     ↓
# Weighted Sum (Σ)
#     ↓
# Activation Function
#     ↓
# Output (0 or 1)
```

| Component           | Symbol | Work                      |
| ------------------- | ------ | ------------------------- |
| Input               | x      | Receives data             |
| Weight              | w      | Importance of input       |
| Bias                | b      | Adjusts decision boundary |
| Summation           | Σ      | Calculates weighted sum   |
| Activation Function | f(z)   | Makes decision            |
