# Stemming and Lemmatization
- Stemming is the process of reducing a word to its root form by removing prefixes or suffixes.
- It is a rule-based approach and does not consider grammar or dictionary meaning.

# Stemming

```python
from nltk.stem import PorterStemmer

stemmer =  PorterStemmer()
```

```python
stemmer.stem("playing")
```

```python
stemmer.stem("running")
```

```python
stemmer.stem("runnings")
```

# Lemmatization
- Lemmatization is the process of converting a word into its meaningful dictionary root form, known as a lemma

- It uses:

- Dictionary lookup
- Grammar rules
- Part-of-Speech (POS) information

```python
import nltk
nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
lemmatizer.lemmatize("running", pos="v")
```

```python
lemmatizer.lemmatize("better", pos="a")
```

```python
lemmatizer.lemmatize("studies", pos="v")
```

# Stemming vs Lemmatization

| Feature           | Stemming               | Lemmatization                  |
| ----------------- | ---------------------- | ------------------------------ |
| Approach          | Rule-Based             | Dictionary-Based               |
| Speed             | Faster                 | Slower                         |
| Accuracy          | Lower                  | Higher                         |
| Grammar Awareness | No                     | Yes                            |
| Output            | May not be a real word | Always a valid dictionary word |
| Example           | studies → studi        | studies → study                |


#  Stemming is the process of reducing words to their root form by removing prefixes and suffixes using predefined rules.

# Lemmatization is the process of converting a word into its meaningful dictionary root form using vocabulary and grammatical analysis.

# Stemming = Cut the word
# Lemmatization = Find the dictionary root word

# 

| Technique         | Advantages                                                                                                               | Disadvantages                                                                                                             |
| ----------------- | ------------------------------------------------------------------------------------------------------------------------ | ------------------------------------------------------------------------------------------------------------------------- |
| **Stemming**      | ✅ Faster processing<br>✅ Uses less memory<br>✅ Easy to implement<br>✅ Good for search engines and basic NLP tasks        | ❌ Less accurate<br>❌ May produce non-dictionary words<br>❌ Ignores grammar and context<br>❌ Can change word meaning       |
| **Lemmatization** | ✅ More accurate<br>✅ Produces valid dictionary words<br>✅ Considers grammar and context<br>✅ Better meaning preservation | ❌ Slower than stemming<br>❌ Requires dictionary lookup<br>❌ More computationally expensive<br>❌ More complex to implement |

| ------------- | -------- | ------------- |

| studies       | studi    | study         |
| running       | run      | run           |
| better        | better   | good          |
