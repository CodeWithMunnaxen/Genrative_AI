# Text Normalization
- Text Normalization is the process of converting text into a standard and consistent format so that machines can process it more effectively.
- Text Normalization makes text clean, uniform, and machine-friendly

# 

# Why Text Normalization?
# Advantages

- Reduces data size

- Improves model accuracy

- Removes noise

- Faster processing

- Creates consistent data

# Disadvantages
- Important information may be lost
- Meaning can change

# Text Segmentation

- Text Segmentation is the process of dividing text into meaningful units such as:

- Paragraphs
- Sentences
- Words
- Topics

# In simple words:

- Segmentation = Breaking large text into smaller meaningful parts.

# Sentence Segmentation

```python
from nltk.tokenize import sent_tokenize

text = "Munna loves AI. Munna learns Python."

sentences = sent_tokenize(text)
```

```python
sentences
```

# Word Segmentation

```python
from nltk.tokenize import word_tokenize

text = "Munna loves AI"

words = word_tokenize(text)
```

```python
words
```

| Feature  | Text Normalization | Text Segmentation |
| -------- | ------------------ | ----------------- |
| Purpose  | Standardize text   | Divide text       |
| Output   | Clean text         | Smaller units     |
| Example  | LOVES → loves      | Sentence → Words  |
| Goal     | Consistency        | Structure         |
| Used For | Cleaning           | Splitting         |


# Text Normalization = Clean and standardize text.

# Text Segmentation = Split text into meaningful parts
