# What are Stop Words
- Stop Words are very common words that appear frequently in a language but usually do not add significant meaning to the sentence

- These words are often removed during text preprocessing to reduce noise and focus on important words

```python
# example:

# is
# am
# are
# was
# were
# the
# a
# an
# and
# or
# but
# in
# on
# at
# to
# of
# for
```

```python
import nltk

nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
text = "Munna is learning AI in Delhi"
words = text.split()

stop_words = set(stopwords.words('english'))

filtered_words = []

for word in words:
    if word.lower() not in stop_words:
        filtered_words.append(word)
```
