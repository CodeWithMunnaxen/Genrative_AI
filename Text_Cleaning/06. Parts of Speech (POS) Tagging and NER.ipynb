# Parts of Speech (POS) Tagging and NER
- th POS Tagging and NER (Named Entity Recognition) are important Natural Language Processing (NLP) techniques. They help Generative AI models understand text before generating responses.

# 1. Parts of Speech (POS) Tagging
# What is POS Tagging?

- POS Tagging is the process of identifying the grammatical role of each word in a sentence.

# It tells whether a word is:

- Noun
- Verb
- Adjective
- Adverb
- Pronoun
- Preposition
- Conjunction
- Interjection

# 

# 

| Feature  | POS Tagging            | NER                       |
| -------- | ---------------------- | ------------------------- |
| Purpose  | Grammar Identification | Entity Identification     |
| Focus    | Word Role              | Real-world Objects        |
| Output   | Noun, Verb, Adjective  | Person, Location, Company |
| Example  | Running → Verb         | Google → Organization     |
| Used For | Grammar Understanding  | Information Extraction    |

```python
from nltk.tokenize import word_tokenize
from nltk import pos_tag

text = "Munna is learning Generative AI."

words = word_tokenize(text)

pos_tags = pos_tag(words)

print("Word\t\tPOS Tag")
print("-" * 25)

for word, tag in pos_tags:
    print(f"{word}\t\t{tag}")
```

```python
from nltk.tokenize import word_tokenize
from nltk import pos_tag

text = "Munna is learning Generative AI."

words = word_tokenize(text)

tags = pos_tag(words)

tag_meaning = {
    "NN":"Noun",
    "NNP":"Proper Noun",
    "VB":"Verb",
    "VBG":"Verb Continuous",
    "VBZ":"Verb Singular",
    "JJ":"Adjective"
}

for word, tag in tags:
    meaning = tag_meaning.get(tag, "Other")
    print(word, "->", tag, "->", meaning)
```

```python
# ==========================
# POS Tagging + NER Example
# ==========================

# Install (Run Once in Terminal)
# pip install nltk spacy
# python -m spacy download en_core_web_sm

import nltk
import spacy
from nltk.tokenize import word_tokenize
from nltk import pos_tag

# Download NLTK Resources (Run Once)
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')

# Load SpaCy Model
nlp = spacy.load("en_core_web_sm")

# Input Text
text = "Munna works at Google in Bangalore and is learning Generative AI."

# ==========================
# POS TAGGING
# ==========================

print("=== POS TAGGING ===")

words = word_tokenize(text)
pos_tags = pos_tag(words)

for word, tag in pos_tags:
    print(f"{word:<15} -> {tag}")

# ==========================
# NER
# ==========================

print("\n=== NAMED ENTITY RECOGNITION (NER) ===")

doc = nlp(text)

for ent in doc.ents:
    print(f"{ent.text:<15} -> {ent.label_}")
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "Elon Musk founded Tesla in California in 2003."

doc = nlp(text)

for ent in doc.ents:
    print(ent.text, "->", ent.label_)
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "Elon Musk founded Tesla in California in 2003."

doc = nlp(text)

print("Entity\t\tType")
print("-" * 30)

for ent in doc.ents:
    print(f"{ent.text}\t\t{ent.label_}")
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "Elon Musk founded Tesla in California in 2003."

doc = nlp(text)

label_meaning = {
    "PERSON":"Person Name",
    "ORG":"Organization",
    "GPE":"Country/City/State",
    "DATE":"Date",
    "MONEY":"Money"
}

for ent in doc.ents:
    print(
        ent.text,
        "->",
        ent.label_,
        "->",
        label_meaning.get(ent.label_, "Other")
    )
```
