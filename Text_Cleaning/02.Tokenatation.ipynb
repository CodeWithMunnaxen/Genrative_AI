# Tokenization
- Tokenization is the process of breaking a piece of text into smaller units called tokens

- A token can be:

- A word
- A sentence
- A character
- A subword

# I love Generative AI

- After Tokenization:

- ["I", "love", "Generative", "AI"]

# Why Do We Need Tokenization?

- Computers don't understand text directly.

- They understand:

- 0s and 1s
- Numbers
- Vectors

- So before converting text into numbers, we first break it into tokens.

# Types of Tokenization

# 1. Sentence Tokenization
- Breaks a paragraph into sentences.

```python
# example : I love AI. It is amazing. NLP is interesting.

#     Output : 

#            [
#  "I love AI.",
#  "It is amazing.",
#  "NLP is interesting."
# ]
```

# Use Cases
- Summarization
- Translation
- Chatbots

# 2. Word Tokenization

- Breaks a sentence into words.

# 3. Character Tokenization

- Breaks text into characters.

# 4. Subword Tokenization

- Most important for LLMs.

# Method 2: NLTK

```python
import nltk


from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('punkt_tab')

sentence = "I love AI"
```

```python
word_tokenize(sentence)
```

```python
from nltk.tokenize import word_tokenize

sentence = "I love AI"

tokens = word_tokenize(sentence)

print(tokens)
```

# Advantages of Tokenization
- Makes text machine-readable
- Reduces complexity
- Helps in NLP preprocessing
- Enables embeddings
- Required for LLMs
- Improves model performance
